```python
#subplot 1
mat.subplot(3,3,1)
#mat.title('original_price')
sea.distplot(df['original_price'])

#subplot 2
mat.subplot(3,3,2)
#mat.tile('discounted_price')
sea.distplot(df['discounted_price'])

#subplot 3
mat.subplot(3,3,3)
#mat.title('reviews')
sea.distplot(df['reviews'])

mat.tight_layout()
mat.show()
```

```python
df1=df[(df.original_price>10000)&(df.discounted_price>1000)]
sea.jointplot(x='original_price', y='discounted_price',data=df1)
mat.show()
```

```python
df1=df[(df.original_price>10000)&(df.discounted_price>1000)]
sea.jointplot(x='original_price', y='discounted_price',data=df1,kind="hex",color="r")
mat.show()
```
